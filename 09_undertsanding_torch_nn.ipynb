### Module


```python
import torch.nn as nn
```

```python
from IPython.display import Image
Image(filename='layers.png')
```

```python
class Model(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        ## Layers
        self.layer1 = nn.Linear(128,32)
        self.layer2 = nn.Linear(32,16)
        self.layer3 = nn.Linear(16,1)
    
    def forward(self,features):
        #(32,128)
        x = self.layer1(features)
        print(x.shape)
        #(32,32)
        x = self.layer2(x)
        x = self.layer3(x)
        # (32,1)
        return x
```

```python
model = Model()
```

```python
features = torch.rand((2,128))
print(features)
```

```python
model(features)
```

```python
features.device
```

```python
## transfer the Features device 'cpu' to 'cuda'
features = features.to('cuda')
model= Model()
model.to('cuda')
model(features)
```

### Sequencial

```python
class Model(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        ## Sequencial
        self.base = nn.Sequential(
            nn.Linear(128,32),
            nn.Linear(32,16),
            nn.Linear(16,1)
        )
        
    
    def forward(self,features):
        #(32,128)
        x = self.base(features)
        print(x.shape)
        # (32,1)
        return x
```

```python
model = Model()
model(features)
```
